# Basic Functions

1. Get transcript(s)
2. List available transcripts
2. Translate transcript

Bill Gates: The next outbreak? We’re not ready | TED

```python
# 1. GET TRANSCRIPTS

# copy video Id. This comes after an = sign.
# if ID starts with a hyphen, then use \ to concatenate the ID.

# 1.1 get_transcript('video_id'): returns a dictionary with text, start, and duraion parameters.

quickTranscript = YouTubeTranscriptApi.get_transcript('6Af6b_wyiwI&ab_channel=TED')
quickTranscript
```

```python
# 1.2 get_transcript('video_id', languages={'en'})
#     Returns a dictionary with text, start, and duraion parameters in the specified language.

esTranscript = YouTubeTranscriptApi.get_transcript('6Af6b_wyiwI&ab_channel=TED', languages={'es', 'en'})
esTranscript
```

```python
# 1.3 get_transcripts(['video_id1', 'id2'])
#     returns a dictionary with text, start, and duraion parameters.
# vid2: What Happens In One Minute?
# https://www.youtube.com/watch?v=zhWDdy_5v2w&ab_channel=AsapSCIENCE

twoTranscripts = YouTubeTranscriptApi.get_transcripts(['6Af6b_wyiwI&ab_channel=TED', 'zhWDdy_5v2w&ab_channel=AsapSCIENCE'])
twoTranscripts
```

```python
# 2. LIST OF AVAILABLE TRANSCRITS

# 2.1 list_transcripts(video_id):
#     returns an object with of all transcripts available.
transcript_list = YouTubeTranscriptApi.list_transcripts('6Af6b_wyiwI&ab_channel=TED')
print(transcript_list)
```

```python
# find specific transcript availability
transcript = transcript_list.find_transcript(['de', 'en'])
print(transcript)
```

```python
# filter for manually created transcripts
manual_transcripts = transcript_list.find_manually_created_transcript(['es', 'en'])
print(manual_transcripts)

# or automatically generated ones
# auto_transcripts = transcript_list.find_generated_transcript(['none'])
# print(auto_transcripts)
```

```python
# data fetch after found transcript
transcript.fetch()
```

```python
# 3. TRANSLATE TRANSCRIPT

# 3.1 translate('language')
#     returns a translated object with text, start, and duraion parameters.
```

```python
base_transcript = transcript_list.find_transcript(['en', 'fr'])
translated_transcript = base_transcript.translate('ay')
print(translated_transcript.fetch())
```

# Export transcripts:

```python
esTranscript = YouTubeTranscriptApi.get_transcript('6Af6b_wyiwI&ab_channel=TED', languages={'es', 'en'})
esTranscript
```

```python
#append all text elements from the json file to get rid of the start and duration parameters.
outLs = []

for i in esTranscript:
  outTxt = (i["text"])
  outLs.append(outTxt)

  with open("op.txt", "a") as opf:
    opf.write(outTxt + "\n")

outLs
```

```python
#append all text elements from the json file to get rid of the start and duration parameters.
outLs = []

for i in translated_transcript.fetch():
  outTxt = (i["text"])
  outLs.append(outTxt)

  with open("op.txt", "a") as opf:
    opf.write(outTxt + "\n")

outLs
```

# Sentiment Analysis on transcripts

```python
from textblob import TextBlob
```

```python
tf = YouTubeTranscriptApi.get_transcript('6Af6b_wyiwI&ab_channel=TED')

outLs = []

for i in tf:
  outTxt = (i["text"])
  outLs.append(outTxt)

  with open("op.txt", "a") as opf:
    opf.write(outTxt + "\n")

outLs
```

```python
#Traversing the transcript list
tdata = ' '
for x in outLs:
  tdata += ' ' + x

print(tdata)
```

```python
yt_vid_transcript = TextBlob(tdata)
```
